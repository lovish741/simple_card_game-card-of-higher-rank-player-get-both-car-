```python
class card:
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
    def __str__(self):
        return (f'{self.rank} of {self.suit}')
```

```python
suit=['Heart','Spade','Club','Diamond']
rank=['two','three','four','five','six','seven','eight','nine','ten','jack','queen','king','ace']
val={'two':2,'three':3,'four':4,'five':5,'six':6,'seven':7,'eight':8,'nine':9,'ten':10,'jack':11,'queen':12,'king':13,'ace':14}
val['two']
```

```python
class deck:
    def __init__(self):
        self.allCard=[]
        for s in suit:
            for r in rank:
                tmp=card(s,r)
                self.allCard.append(tmp)
```

```python
d=deck()
```

```python
class player:
    def __init__(self,name):
        self.playingCards=[]
        self.name=name
    def __str__(self):
        return (f'Player {self.name} have {len(self.playingCards)} cards!!!')
    def __len__(self):
        return len(self.playingCards)
    def add_card(self,card):
        self.playingCards.append(card)
    def add_multiplecard(self,card):
        self.playingCards.extend(card)
    def remove_card(self):
        return self.playingCards.pop(0)
```

```python
p1=player('lovish')
p2=player('charchit')
```

```python
#game begin
player1_card=[]
player2_card=[]
round=0
while True:
    round+=1
    print(f'Round {round}')
    if len(p1)==0:
        print('Player 1  lose. Player 2 Wins!!!')
        break
    if len(p2)==0:
        print('Player 2 lose. Player 1 Wins!!!')
        break
    player1_current_card=p1.remove_card()
    player2_current_card=p2.remove_card()
    player1_card.append(player1_current_card)
    player2_card.append(player2_current_card)
    if val[player1_current_card.rank]>val[player2_current_card.rank]:
        p1.add_multiplecard(player1_card+player2_card)
        print(f'card added to player1 {len(player1_card+player2_card)}')
        player1_card=[]
        player2_card=[]
    elif val[player1_current_card.rank]<val[player2_current_card.rank]:
        p2.add_multiplecard(player2_card+player1_card)
        print(f'card added to player2 {len(player1_card+player2_card)}')
        player1_card=[]
        player2_card=[]
    else:
        print('war')
```
